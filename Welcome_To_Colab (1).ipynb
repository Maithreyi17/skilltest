```
#3a
import cmath
import pandas as pd
def imp(c,v,f):
  w=2*cmath.pi*f
  if c=='R':
    return complex(v,0)
  elif c=='L':
    return complex(0,w*v)
  elif c=='C':
    return complex(0,-1/(w*v))
  else:
    return 'Invalid component'
def tot(df,config):
  if config=='Series':
    return df['Impedance'].sum()
  else:
    return 1 / sum(1 / z for z in df['Impedance'])
f=float(input("Enter Frequency:"))
config=input("Enter the configuration(Series/Parallel):")
n=int(input('No of components:'))
d=[]
for i in range(n):
  c=input("Enter the component(R/L/C):").upper()
  v=float(input("Enter the value:"))
  z=imp(c,v,f)
  d.append({'Component':f'{c}{i+1}','Type':c,'Value': v, 'Impedance': z})
df=pd.DataFrame(d)
pd.set_option("display.precision",2)
print(df)
t=tot(df,config)
print(f'Total Impedance:{t:3.2f}')
```

```
#1a
scores = [85, 42, 77, 90, 68, 55, 33, 89, 92, 47, 73, 50, 60, 95, 39, 81, 70, 66, 58, 76]

# 1. Overall class performance
average_score = sum(scores) / len(scores)
highest_score = max(scores)
lowest_score = min(scores)

print("Class Performance")
print(f"Average Score: {average_score:.2f}")
print(f"Highest Score: {highest_score}")
print(f"Lowest Score: {lowest_score}")

# 2. Pass/Fail Classification (Pass mark is 50)
passed_students = [score for score in scores if score >= 50]
failed_students = [score for score in scores if score < 50]

print("\nPass/Fail Classification")
print(f"Number of Passed Students: {len(passed_students)}")
print(f"Number of Failed Students: {len(failed_students)}")
print(f"Passed Scores: {passed_students}")
print(f"Failed Scores: {failed_students}")

# 3. Grading on a Curve (using predefined thresholds)

def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

grades = [assign_grade(score) for score in scores]

print("\n Grading")
for i, (score, grade) in enumerate(zip(scores, grades), start=1):
    print(f"Student {i}: Score = {score}, Grade = {grade}")
```
